By using the methods of beautiful soup to arrive at the desired location of the wanted data on the webpage.

```python
def myfun(urlbase):
    #urlbase = "https://www.the-numbers.com/person/25200401-D-J-Caruso#tab=summary"
    want1= requests.get(urlbase)
    want2 =want1.text
    want3 =BeautifulSoup(want2,'lxml-xml')
    want4=want3.find_all(href=re.compile("box-office-star-record"))
    result1=set([want4[i].text for x in want4 for i in range(0,len(want4))])
    want5=want3.find_all(align=re.compile("right"))
    result2=[want5[i].text for x in want4 for i in range(0,len(want5))][0:len(result1)]
    '''here, result1 and 2 is actually not useful anymore since we can get the whole table in the below functions to obtain the all-aspects data from the websites'''
    want6=want3.find_all(name="td",attrs={'class':'alnright'})
    result3=[want6[i].text for x in want6 for i in range(1,4)][0:3]
    return result3
```

```python
def myfun2(url):
    '''this function is used for getting the data for specific study for director'''
    #url = "https://www.the-numbers.com/person/155860401-Donnie-Yen#tab=summary"
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'lxml')
    tables = soup.select('table')
    
    m=pd.read_html(tables[2].prettify())[0][1:]
    return m
```
