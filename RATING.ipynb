```python
def myfun(urlbase):
    #urlbase = "https://www.the-numbers.com/person/25200401-D-J-Caruso#tab=summary"
    want1= requests.get(urlbase)
    want2 =want1.text
    want3 =BeautifulSoup(want2,'lxml-xml')
    want4=want3.find_all(name='span',attrs={'itemprop':'ratingValue'})
    want5=want4[0].text
    want6=want3.find_all(name='span',attrs={'class':'subText'})
    want7=want6[2].text.strip()
    want8=re.split(r"\n",want7)
    return want8,want5
```

```python
link=["http://www.imdb.com/title/tt1293847/",
      "http://www.imdb.com/title/tt1981128/",
      "http://www.imdb.com/title/tt4630562/",
      "http://www.imdb.com/title/tt2345759/",
      "http://www.imdb.com/title/tt3371366/?ref_=fn_al_tt_1",
      "http://www.imdb.com/title/tt3731562/?ref_=fn_al_tt_1",
      "http://www.imdb.com/title/tt1753383/?ref_=fn_al_tt_1"]
Popularity=[]
Rate=[]
for l in link:
    x=myfun(l)
    p=x[0][0]
    r=x[1]
    Popularity.append(p)
    Rate.append(r)   
```

```python
Name=["xXx: Return of Xander Cage",
      "Geostorm",
      "The Fate of the Furious",
      "The Mummy",
      "Transformers: The Last Knight",
      "Kong: Skull Island",
      "A Dog’s Purpose"]
dict={'Name':Name,
      'IMDbPopularity':Popularity,
      'IMDbRate':Rate}
Rating=pd.DataFrame(dict)
Rating=Rating[["Name",'IMDbPopularity','IMDbRate']]
Rating
```
